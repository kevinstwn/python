```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 100)
y = x-(x**3 / 6)
plt.plot(x, y, label = 'Fourth Order')
plt.plot(x, np.sin(x), 'k', label = 'Analytic')

plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 100)
y = 1 - (x**2 / 2) + (x**4 / 24)
plt.plot(x, y, label = 'Fourth Order')
plt.plot(x, np.cos(x), 'k', label = 'Analytic')

plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 100)
y = x - (4*x**3 / 6)
plt.plot(x, y, label = 'Fourth Order')
plt.plot(x, np.sin(x)*np.cos(x), 'k', label = 'Analytic')

plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```
